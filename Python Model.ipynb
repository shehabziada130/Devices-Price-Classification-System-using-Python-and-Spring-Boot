##### Split the data

```python
X=train_data.drop(['price_range'],axis='columns')
y=train_data['price_range']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split (X, y, test_size=0.25)
```

```python
X_train.shape,X_val.shape
```

##### Scale The Data Using Standard scaler

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_val_scaled=scaler.transform(X_val)
```

```python
X_train_scaled[0:5]
```

### Grid Search Over The models 

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model_params={
    'svm':{
        'model':SVC(),
        'params':{
                'kernel': ['linear','poly'],
                'C': [1, 5, 10]
        }
  
    },
    'rfc':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    },
   'dtc':{
       'model':DecisionTreeClassifier(),
       'params':{
            'max_depth':[3,5,10],
            'criterion' : ['gini','entropy']
       }
   } 
    
}
```

```python
scores=[]
for model_name, mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(X_train_scaled,y_train)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
scores
```

```python
model=SVC(C=10,kernel ='linear')
model.fit(X_train_scaled,y_train)
```

```python
model.score(X_val_scaled,y_val)
```

#### SVM is best Model, Now we will Grid Search Over SVM parameters

```python
svm_params={
    'SVM':{
        'model':SVC(),
        'params':{
                'kernel': ['linear','poly','sigmoid'],
                'C': [2,3,5,10,20],
        }
  
    },
}
```

```python
scores=[]
for model_name, mp in svm_params.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(X_train_scaled,y_train)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
scores
```

```python
model=SVC(C=20,kernel ='linear')
model.fit(X_train_scaled,y_train)
```

```python
model.score(X_val_scaled,y_val)
```

### Make The Pipeline

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
 ('std_scaler', StandardScaler()),
 ('Model' , SVC(C=20,kernel ='linear')),
])
```

```python
pipeline.fit(X_train,y_train)
```

```python
pipeline.score(X_train,y_train)
```

### Model Testing 

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred=pipeline.predict(X_val)
accuracy = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred, average='weighted')
recall = recall_score(y_val, y_pred, average='weighted')
f1 = f1_score(y_val, y_pred, average='weighted')
```

```python
data={
    'Metric':['Accuracy','Precision', 'Recall', 'F1' ],
    'Value':[accuracy,precision,recall,f1]
}
metrics_df=pd.DataFrame(data)
```

##### Prediction Function 

```python
def predict(params_dict):
    x = np.zeros(len(X.columns))

    for param_name, param_value in params_dict.items():
        index = X.columns.get_loc(param_name)
        if index >= 0:
            x[index] = param_value
        else:
            print(f"Ignoring unknown parameter: {param_name}")

    return pipeline.predict([x])[0]
```

```python
params =X_val.iloc[0]

predict(params)
```
